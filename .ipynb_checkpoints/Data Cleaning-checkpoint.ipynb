```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('test_data.csv')
df
```

```python
df.describe()
```

# Filtering a DataFrame

```python
age_filtered_df = df[df['Age'] > 18]
age_filtered_df
```

```python
score_filtered_df = df[(df['Score'] > 50) & (df['Score'] < 100)]
score_filtered_df
```

# Modifying a DataFrame

```python
df_cleaned = df[df['Score'] >= 0]
df_cleaned
```

```python
mean_age = df_cleaned['Age'].mean()
df_cleaned.loc[:,'Age'] = df_cleaned['Age'].fillna(mean_age)
df_cleaned
```

```python
mean_score = df_cleaned['Score'].mean()
df_cleaned.loc[:,'Score'] = df_cleaned['Score'].replace(0, mean_score)
df_cleaned
```

```python
df_cleaned = df_cleaned.dropna()
df_cleaned
```

```python
df_cleaned = df_cleaned.set_index('ID')
df_cleaned
```
