```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Creating Basic Plots

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

# Styling Plots

```python
plt.plot(x, y, linestyle='--', color='red')
plt.show()
```

```python
plt.plot(x, y, linestyle='--', color='red', marker='s', markersize=10, markerfacecolor='green')
plt.show()
```

```python
plt.plot(x, y, linestyle='--', color='red', marker='^', markersize=10, markerfacecolor='blue')
plt.show()
```

```python
plt.plot(x, y, linestyle='--', color='red', marker='D', markersize=10, markerfacecolor='purple')
plt.show()
```

# Advanced Plotting Techniques

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [23, 17, 35, 29]
```

```python
plt.bar(products, sales, color='skyblue', width=0.5)
plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 100  
colors = np.random.rand(50) 
```

```python
plt.scatter(x, y, s=sizes, c=colors, cmap='viridis')
plt.colorbar()  # Adds a color scale legend
plt.title("Scatter Plot with Variable Marker Size and Color")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

# subplots

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
data = np.random.randn(1000)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
# line chart
ax1.plot(x, y, color='blue')
ax1.set_title("Sine Wave")
ax1.set_xlabel("X values")
ax1.set_ylabel("Sin(X)")
# histogram
ax2.hist(data, bins=30, color='purple')
ax2.set_title("Histogram of Random Data")
ax2.set_xlabel("Value")
ax2.set_ylabel("Frequency")
plt.tight_layout()  
plt.show()
```
